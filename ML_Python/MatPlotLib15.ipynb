```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-3,3,0.001)
print (x)
print (norm.pdf(x))
plt.plot(x, norm.pdf(x))
plt.show()
```

# Mutiple plots on one graph

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

# Box and Whisker Plot
* useful for visulizing the spread and skew of data
* the red line represents the median of the data, and the box reperesents the bounds of the 1st and 3rd quartiles
* So, half of the data exists within the box
* the dotted-line 'whiskers' indicate the ragne of the data, except for outliers, which are plotted outside the whiskers, outliers are 1.5X or more the interquartile range


```python
# this example below creates uniformly distributed random unmbers between -40 and 60, plus a few outliers about 100 and below -100
```

```python
uniformSkewed = np.random.rand(100)*100-40
high_outliers = np.random.rand(10)*50 +100
low_outliers = np.random.rand(10)* -50 -100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```
